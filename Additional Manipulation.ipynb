```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline

# Reading in the data
meta = pd.read_excel('C://Users/tpensaert/Documents/Thesis/meta_data_final.xlsx', index=False)

# Converting date type + adding additional columns
meta['Publication Date'] = pd.to_datetime(meta['Publication Date'])
meta['Year'] = pd.DatetimeIndex(meta['Publication Date']).year
meta['Rank'] = meta.groupby(['Year', 'Issue'])['First page'].rank(ascending=True)

meta.head()
```

```python
meta.to_excel("meta_data_final2.xlsx")
```

```python
def create_grid(data, xaxis, yaxis, color, dimension):
    
    # Create a grid-plot, showing 3D spread of articles
    sns.lmplot(xaxis, yaxis, data=data, 
               hue=color, fit_reg=False, 
               col=dimension, col_wrap=2)
```

```python
meta = pd.read_excel('C://Users/tpensaert/Documents/Thesis/meta_data_final2.xlsx', index=False)
create_grid(meta, "Subjectivity", "Polarity", "Year", "Pvalue distribution")
```
